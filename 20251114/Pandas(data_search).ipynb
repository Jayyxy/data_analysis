## 1. 데이터 생성  

```python
# (1) 딕셔너리로 생성
# key 컬럼 value 해당 컬럼의 값 
data1 = {
    '이름' : ['철수','영희','민수'],
    '나이' : [25,30,22],
    '도시' : ['서울','부산','대구']
}
```

```python
# (2) 중첩 리스트(2차원 리스트) 생성

data2 = [
    ['철수',25,'서울'],
    ['영희',30,'부산'],
    ['민수',22,'대구']
]

# 컬럼명 별도 지정 
df2 = pd.DataFrame(data2, columns=['이름','나이','도시'])
print((df2))
```
