# 데이터 변화의 필요성 
- 데이터는 항상 원하는 타입으로 되어있지 않음 

df["컬러명"].astype(변환할타입)

```python
import pandas as pd 

df = pd.read_csv('data/seoul_park.csv') # .csv파일
```

```python
# pandas의 메서드는 덮의쓰기 작업이 필요 
# df["컬러명"].astype(변환할 타입)

df["어른"] = df["어른"].astype(int)
```

```python
df["유료합계"] = pd.to_numeric(df["유료합계"])  # 숫자형 타입으로 변환 
```

```python
df["날짜"] = pd.to_datetime(df["날짜"])  # 날짜형 타입으로 변환 
```

### 시간 속성 : dt접근자 사용



```python
df["연"] = df["날짜"].dt.year

df["월"] = df["날짜"].dt.month

df["일"] = df["날짜"].dt.day

df["요일"] = df["날짜"].dt.dayofweek # 요일은 숫자로 표현 

# map()을 이용하여 시리즈 값 변환
week = {0: "월",1 : "화",2 : "수", 3 : "목" , 4 : "금" , 5 : "토", 6 : "일"} 
df["요일"] = df["요일"].map(week)
```

# 데이터 값 함수 적용하여 변환하기 : apply() 
    df.apply(함수, axis=0 또는 1)
    df[“특정 컬럼”].apply(함수, axis=0 또는 1)

```python
def weather(e):
    if e == "눈" or e == "비" :
        return "눈/비"
    else :
        return e 
    
df["날씨"] = df["날씨"].apply(weather) # 날씨 컬럼의 값을 weather함수에 넣어 return한 값을 할당 
```
