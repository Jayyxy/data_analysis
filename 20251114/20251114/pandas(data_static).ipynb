# 통계 기반 데이터 요약하기 
- 방대한 데이터를 한 눈에 파악하는 것은 불가능 > 먼저 데이터의 전반적인 형태를 요약
- 정보 미리 파악하여 추후 분석 방향 판단 가능 

## 먼저 요약할 데이터 형태
1. 대표값 : 평균, 중앙값, 최댓값, 최솟값
2. 빈도 분포(범주형)
3. 분산, 표준편차(수치형)
4. 결측치, 이상치의 존재 여부 

```python
import pandas as pd 

data= {'키' : [155,167,175],
      '몸무게' : [43,56,57]} 


df = pd.DataFrame(data)
```

```python
df.info()
```

```python
# (1) 대표값
df.mean() # 전체에 대하여 평균 
df["키"].mean() # 평균

df.median()

df.mode()

```

```python
df.std()

df.max()

```

```python
print(df.quantile(0.25)) # 1분위수
print(df.quantile(0.75)) # 3분위수 
```

```python
df.sum() 
df.cumsum() #누적합 
```
